**Normalize the dataset using the MinMaxScaler after removing the following columns: [“date”, “lights”]. The target variable is “Appliances”. Use a 70-30 train-test set split with a random state of 42 (for reproducibility). Run a multiple linear regression using the training set and evaluate your model on the test set. Answer the following questions:**

**What is the Mean Absolute Error (in two decimal places)**

```python
# dropping the date and light column

df = df.drop(['date', 'lights'], axis=1)
```

```python
# Normalizing the data

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
energy_df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
```

```python
X = energy_df.drop('Appliances', axis = 1)
y = energy_df['Appliances']
```

```python
energy_df.head()
```

```python
# Model Training

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

**From the dataset, fit a linear model on the relationship between the temperature in the living room in Celsius (x = T2) and the temperature outside the building (y = T6). What is the R^2 value in two d.p.?**

```python
# We had to reshape our data because we have just a feature

t2 = np.array(energy_df['T2'])
t6 = np.array(energy_df['T6'])
t2 = t2.reshape(-1, 1)
t2_train, t2_test, t6_train, t6_test = train_test_split(t2, t6, test_size=0.3, random_state=42)
```

```python
#Using Linear regression model

from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(t2_train, t6_train)
t6_pred = model.predict(t2_test)
```

```python
# R2 Score

from sklearn.metrics import r2_score
r2 = r2_score(t6_test, t6_pred)
print('The r2 score is', (round(r2,2)))
```

```python
reg = LinearRegression()
reg.fit(X_train, y_train)
y_pred = reg.predict(X_test)
```

```python
# Mean absolute error (MAE)

from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, y_pred)
print('The mean absolute error is', (round(mae, 2)))
```

**What is the Residual Sum of Squares (in two decimal places)?**

```python
# Residual sum of squares

rss = np.sum(np.square(y_test - y_pred))
print('The residual sum of square is', (round(rss,2)))
```

**What is the Root Mean Squared Error (in three decimal places)?**

```python
# Root Mean Squared Error

from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
```

```python
rmse = np.sqrt(mse)
print('The root mean square error is', (round(rmse, 3)))
```

**What is the Coefficient of Determination (in two decimal places)?**

```python
# Coefficient of Determination

cod = r2_score(y_test, y_pred)
print('The coefficient of determination is', (round(cod, 2)))
```

**Train a ridge regression model with an alpha value of 0.4. Is there any change to the root mean squared error (RMSE) when evaluated on the test set?**

```python
from sklearn.linear_model import Ridge

# define model
model_ridge = Ridge(alpha=0.4)
# fit model
model_ridge.fit(X_train, y_train)
ridge_pred = model_ridge.predict(X_test)
```

```python
rmse = np.sqrt(mean_squared_error(y_test, ridge_pred))
print('The root mean square error is', (round(rmse, 3)))
```
